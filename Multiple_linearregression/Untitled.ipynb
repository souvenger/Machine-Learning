```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('car_data.csv')
inp=pd.read_csv('Input.csv')
```

```python
data.columns
data
```

```python
y=data['Price']
data1=data
data1.drop(['Color','Model','Price'],axis=1,inplace=True)
data1['Fuel_Type']=data1['Fuel_Type'].apply(lambda x:1 if (x=='Petrol') else 0)
```

```python
data1
x_train,x_test,y_train,y_test=train_test_split(data1,y,test_size=0.1,random_state=42)
```

```python
model=linear_model.LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
mse=metrics.mean_absolute_error(y_test,y_pred)
mse
mae=metrics.mean_squared_error(y_test,y_pred)
mae
```

```python
output=model.predict(inp)
```
