```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data collection and Preprocessing

```python
movies_data=pd.read_csv("/content/sample_data/movies.csv")
```

# New Section

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
#Feature Selection 
features=['genres','keywords','cast','tagline','director','overview']
```

```python
#removing null and replacing missing values
for feature in features:
   movies_data[feature]=movies_data[feature].fillna(' ')
```

```python
new_feature=""
for x in features:
  new_feature+=movies_data[x]
new_feature
```

```python
#using vectorizer converting text data to vectors
vectorizer=TfidfVectorizer()
```

```python
#making vectors
vectors=vectorizer.fit_transform(new_feature)
```

```python
print(vectors)
```

```python
#Getting similarity Score for every movie and compare this score
#to show which movie is similar to which
similarity_score=cosine_similarity(vectors)
print(similarity_score)
```

```python
#ALL in one Frame
#input movie name from user
movie_name=input("Enter ur movie name ")

#creating list of movies from dataset just storing the title
movie=movies_data['title'].tolist()

similar_movies=difflib.get_close_matches(movie_name,movie)

close_matched=similar_movies[0]
close_matched

#find row index for closed_match movie
#so that we can compare with its attributes
idx=movies_data[movies_data.title==close_matched]['index'].values[0]

#list of similar movies
similarity_movie_score=list(enumerate(similarity_score[idx]))

movie_list=sorted(similarity_movie_score,key=lambda x:x[1],reverse=True)

#print name of movies based on index
print("Suggested Movies are : ")
j=0
for i in movie_list:
  suggested_movie=movies_data[movies_data.index==i[0]]['title'].values[0]
  print(suggested_movie)
  j+=1
  if(j==25):
    break
```
